```python
#Subplot
'''In data visualization, a subplot, short for "sub-plot," refers to a feature or capability that allows you to create multiple smaller plots within
a single, larger plot or figure. Subplots are commonly used when you want to display multiple graphs or charts side by side,
making it easier to compare and analyze different aspects of your data within a single visual representation.'''

import numpy as np
import matplotlib.pyplot as plt
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
# Create a subplot with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# Plot Math Scores
axs[0].bar(students, math_scores, color='b')
axs[0].set_title('Math Scores')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Score')
# Plot Science Scores
axs[1].bar(students, science_scores, color='g')
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[1,2,3,4,5,6,7,8,9]
plt.subplot(2,2,1)
plt.plot(x,y,color="g")

plt.subplot(2,2,2)
plt.pie([1],colors="r")

plt.subplot(2,2,3)
a=[10,20,30,40]
plt.pie(a)

plt.subplot(2,2,4)
a1=["r","a","h","u"]
a2=[10,20,30,40]
plt.bar(a1,a2)

plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[1,2,3,4,5,6,7,8,9]
plt.subplot(2,2,1)               #line chart used to track changes over short and long period of time
plt.plot(x,y,color="g")
plt.title("line chart")   

plt.subplot(2,2,2)
plt.pie([1],colors="r")        #pie chart used to organize and show data as a percentage of whole
plt.title("pie chart")

plt.subplot(2,2,3)
a=[10,20,30,40]
plt.pie(a)
plt.title("pie chart")

plt.subplot(2,2,4)
a1=["r","a","h","u"]          #bar chart distribution of data points
a2=[10,20,30,40]
plt.bar(a1,a2)               #hist chart freq distribution
plt.title("bar chart")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,11]
plt.plot(x,y)
plt.title("plot with general text")
plt.text(2,8,"Language", fontsize=10, color="purple", bbox={"facecolor":"pink"})
plt.show()
```

```python
 '''Matplotlib provide several function to customize the apperance of the axes in a plot.
 these includes setting the ticks and limits for both the x axis and y axis .here,you can use xticks,yticks,
 xlim and ylim in matplotlib,along whith some practical examples.""
xticks and yticks:..these functions are used to get or set the current tick location and labels of the xaxis and y axis,
 respectively.
xlim and ylim : these function are used to get or set teh current x axis and y axis limt respectively.'''

x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,12)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xticks([1,2,3,4,5],['one','two','three','four','five'])
plt.yticks([2,4,6,8,10],['two','four','six','eight','ten'])
plt.title("custom tick example")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
