```python
import pandas as pd 
data ={
    "Date ":["2024-01-01","2024-01-01","2024-01-02","2024-01-02","2024-01-03"],
      "Category":["A","B","A","B","A"],
      "Sales":[100,200,150,250,300],
       "Quantity":[10,20,15,25,30]}
df = pd.DataFrame(data)
pivot_table=pd.pivot_table(df,
                          values=["Sales"], #columns to aggregate
                          index=["Date "],    #rows
                          columns=['Category'],  #columns
                          aggfunc={'Sales':'sum'},  #aggregate function 
                          fill_value=0,  #replace NaNe with 0 
                          dropna=False, # Include column with all NaN value
                           margins=True, # Add total 
                           margins_name='Total'  # Name for the totals row / columns 
                          )
pivot_table
```
