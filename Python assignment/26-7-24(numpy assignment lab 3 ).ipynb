```python

#lab3:-----------------------------------------------------

#ques1:-------
'''calculate the total revenue generated by two product categories in a store'''
import numpy as np
category1_revenue = np.array([500, 600, 700, 550])
category2_revenue = np.array([450, 700, 800, 600])
total_revenue_category1 = np.sum(category1_revenue)
total_revenue_category2 = np.sum(category2_revenue)
total_revenue = total_revenue_category1 + total_revenue_category2

print("Total revenue : ", total_revenue)
```

```python
#ques2:-------------
'''calculate the profit made by a company'''

import numpy as np
revenue=np.array([10000,12000,11000,10500])
expenses=np.array([4000,5000,4500,4800])

profit = revenue - expenses

print("Profit for each period:", profit)
total_profit = np.sum(profit)
print("Total profit:", total_profit)
```

```python
#ques3:------------
'''determine which products in a store are out of stock(quantity is 0)'''

import numpy as np
inventory=np.array([10,0,5,0,20,0])
out_of_stock_indices = np.where(inventory == 0)[0]

print("products which are out of stock:", inventory[out_of_stock_indices])
```

```python
#ques4:--------------------
'''Calculate the total cost of items in a shopping cart,considering the quantity and price per item'''

import numpy as np
quantity = np.array([2, 3, 4, 1])
price_per_item = np.array([10.0, 5.0, 8.0, 12.0])

total_cost_per_item = quantity * price_per_item
total_cost = np.sum(total_cost_per_item)

print("Total cost per item:", total_cost_per_item)
print("Total cost of items in the shopping cart:", total_cost)


```
