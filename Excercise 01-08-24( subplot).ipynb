```python
#Excercise Ques.....
'''I have sales data for a company, and I want to create a single figure
with subplots to compare sales figures for different regions over time. Each subplot will
represent a different region.Can you please generate a program on it.'''

import matplotlib.pyplot as plt

# Sample data
dates = ['2024-01', '2024-02', '2024-03', '2024-04']
sales_north = [200, 220, 250, 270]
sales_south = [150, 180, 210, 240]
sales_east = [180, 190, 220, 260]
sales_west = [170, 200, 230, 250]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# North Region
axs[0, 0].plot(dates, sales_north, marker='o')
axs[0, 0].set_title('North Region')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Sales')

# South Region
axs[0, 1].plot(dates, sales_south, marker='o', color='orange')
axs[0, 1].set_title('South Region')
axs[0, 1].set_xlabel('Date')
axs[0, 1].set_ylabel('Sales')

# East Region
axs[1, 0].plot(dates, sales_east, marker='o', color='green')
axs[1, 0].set_title('East Region')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Sales')

# West Region
axs[1, 1].plot(dates, sales_west, marker='o', color='red')
axs[1, 1].set_title('West Region')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Sales')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
